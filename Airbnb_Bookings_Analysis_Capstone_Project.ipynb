# Data Wrangling and Cleaning

```python
#check null 
ar.isnull().sum()
```

```python
#check any duplicate id exits 
boole=ar.duplicated('id').any()
print(boole)
```

-> id has no duplicate value

```python
#check duplicate entries 
boolean = ar.duplicated(subset=['host_id','host_name', 'neighbourhood_group','neighbourhood_group','neighbourhood',	'latitude',	'longitude','room_type']).any()
print(boolean) # True
```

-> From this we can see True that menas our dataframe has duplicate entries

```python
#counting duplicate
ar.duplicated(subset=['host_name', 'neighbourhood_group','neighbourhood_group','neighbourhood',	'latitude',	'longitude','room_type']).sum()
```

```python
# Considering certain columns for dropping duplicates in original dataFrame
ar.drop_duplicates(subset=['id','host_name','neighbourhood_group','room_type'	],inplace=True)
ar.drop_duplicates(subset=['latitude',	'longitude',	'room_type','minimum_nights',	'number_of_reviews',	'last_review',	'reviews_per_month'	],inplace=True)
```

```python
ar.shape
```

```python
# Drop duplicate rows in place
ar.drop_duplicates(keep=False, inplace=False)
```

```python
#check duplicate value exist or not
ar.duplicated().sum()
```

->We do not have any duplicate rows in our dataset.

```python
#Replacing null values in the column reviews_per_month with 0 in the dataset
ar['reviews_per_month'].fillna(0,inplace = True)
#filled host_name and name nan value as # and $ respectively
ar['name'].fillna("$",inplace=True)
ar['host_name'].fillna("#",inplace=True)
```

```python
#check null value exits in last_review
a=pd.isnull(ar['last_review'])
p=a.sum()
print(p)
```

->as last_review contain more than 2000 null values so drop this column.

```python
#drop last_review
ar.drop(['last_review'],axis=1,inplace=True)
```
